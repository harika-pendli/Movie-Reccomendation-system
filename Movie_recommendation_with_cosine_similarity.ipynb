**Movie Recommendation system based on Cosine Similarity**

```python
#import dependencies

import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#load into dataframe
movies_dataset= pd.read_csv('/content/movies.csv')

print(movies_dataset.shape)
movies_dataset.head()
```

```python
features_considered= ['genres','keywords','cast','tagline','director']
```

```python
#fill the nan values in the features selected with empty string

for column in features_considered:
  movies_dataset[column]= movies_dataset[column].fillna('')
```

```python
#merging the features_considered
movies_dataset['combined']= movies_dataset['genres'] + ' '+ movies_dataset['keywords'] + ' ' + movies_dataset['cast']+ movies_dataset['tagline']+ ' '+ movies_dataset['director']
```

```python
#converting text to feature vectors
X=movies_dataset['combined'].values
vectorizer= TfidfVectorizer()
vectorizer.fit(X)

X= vectorizer.transform(X)
print(X)
```

```python
#similarity scores using cosine similarity

similarity = cosine_similarity(X)
print(similarity.shape)
similarity
```

```python
#take input movie
movie_ip = input("Enter movie name:")
```

```python
#finding closest match to the input movie title from our titles in the movie dataset
closest_title= difflib.get_close_matches(movie_ip,movies_dataset['title'])[0]

closest_title
```

```python
index_movie_ip= movies_dataset[movies_dataset.title == closest_title].values[0][0]

print(index_movie_ip)
```

```python
similarity_score = list(enumerate(similarity[index_movie_ip]))
similarity_score
```

```python
sorted_similar_movies= sorted(similarity_score,key= lambda x:x[1], reverse=True)
sorted_similar_movies
```

```python
#therefore similar movies based on the above reverse sorted list

print("Top 10 Recommended movies: ")
rec_movies=[]
for pair in sorted_similar_movies:
  ind= pair[0]
  rec_movies.append(movies_dataset[movies_dataset.index == ind].values[0][7])
print(rec_movies[:10])
```
